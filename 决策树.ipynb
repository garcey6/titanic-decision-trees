### 练习题1

#### 对“正则化超参数”部分的moons数据集进行决策树训练并优化模型

#### 1.通过语句make_moons(n_samples=10000, noise=0.4, random_state=3)生成moons数据集

```python
from sklearn.datasets import make_moons
x,y = make_moons(n_samples = 10000,noise = 0.4,random_state = 3)
```

#### 2.通过train_test_split()将数据集分割为训练集和测试集（如20%作为测试集）

```python
from sklearn.model_selection import train_test_split
#在数据集中选取80%作为训练集，20%作为测试集
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

#### 3.进行交叉验证，并使用网格搜索法寻找最好的超参数值

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier

max_leaf_nodes = [i for i in range(1,20)]
max_depth = [i for i in range(10,50)]

parameters = {'max_leaf_nodes':max_leaf_nodes, 'max_depth':max_depth}
Tree_clf = DecisionTreeClassifier()
grid_search = GridSearchCV(Tree_clf,parameters,verbose=2, cv=3)
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_estimator_
```

#### 4.使用这些超参数训练全部的训练集数据，并在测试集上测量模型的表现。你应该获得大约 85% 到 87% 的准确度

```python
from sklearn.metrics import accuracy_score

#测试集上的准确率
y_pred = grid_search.best_estimator_.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
#训练集上的准确率
y_pred_1 = grid_search.best_estimator_.predict(x_train)
accuracy_score(y_train,y_pred_1)
```

```python
import os
os.environ["PATH"] += os.pathsep + 'C:/Program Files/Graphviz/bin'
```

#### 5.画出最好的超参数值对应的决策树

```python
from sklearn.tree import export_graphviz
import pydotplus

#export_graphviz(grid_search.best_estimator_, out_file='makemoons.dot')
label = ['label1','label2']
data_class = ['0','1']
dot_data = export_graphviz(grid_search.best_estimator_, out_file=None, feature_names=label, class_names=data_class,
                               filled=True, rounded=True, special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data)
graph.write_pdf('makemoons.pdf')
f = open('DecisionTree.png', 'wb')
f.write(graph.create_png())
f.close()
```

### 练习二

#### 泰坦尼克号的沉没是世界上最严重的海难事故之一，造成了大量的人员伤亡。这是一艘号称当时世界上最大的邮轮，船上的人年龄各异，背景不同，有贵族豪门，也有平民旅人，邮轮撞击冰山后，船上的人马上采取措施安排救生艇转移人员，从本次海难中存活下来的，也就是幸存者。应用泰坦尼克数据集，使用决策树方法来预测哪些人可能成为幸存者。

#### 1.使用pandas的read_csv导入文件夹下的数据

```python
import pandas as pd

titanic = pd.read_csv('titanic.csv')
titanic.head()
```

#### 2.解压这些数据集，熟悉它的数据格式

```python
titanic.value_counts()
```

```python
titanic.describe()
```

#### 3.数据预处理，包括选取合适特征、应用合适方式处理缺失值、数值类型转换等操作

```python
#数据集的清洗
from sklearn.impute import SimpleImputer
import numpy as np
imputer = SimpleImputer(strategy="median")

data = titanic.drop(["Name","Sex","Ticket","Cabin","Embarked"], axis=1)

imputer.fit(data)
#发现在数据清洗后仍然存在数据为空的情况
print(np.isnan(data).any())

print('-------------------------------------')
data = data.replace(np.nan,30)
print(np.isnan(data).any())
#此时没有数据为空(np.nan)的情况
```

```python
from sklearn.model_selection import train_test_split

y = data['Survived']
x = data.drop(['Survived','PassengerId'],axis = 1)
```

#### 4.通过train_test_split()将数据集分割为训练集和测试集（如25%作为测试集）

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25,random_state = 42)
```

#### 5.进行交叉验证，并使用网格搜索法寻找最好的超参数值（选择合适的超参数进行优化）

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier

max_leaf_nodes = [i for i in range(1,20)]
max_depth = [i for i in range(10,50)]

parameters = {'max_leaf_nodes':max_leaf_nodes, 'max_depth':[10]}
Tree_clf = DecisionTreeClassifier()
grid_search = GridSearchCV(Tree_clf,parameters,verbose=2, cv=3)
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_estimator_
```

```python
from sklearn.metrics import accuracy_score

#测试集上的准确率
y_pred = grid_search.best_estimator_.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
#训练集上的准确率
y_pred = grid_search.best_estimator_.predict(x_train)
accuracy_score(y_train,y_pred)
```

#### 6.画出最好的超参数值对应的决策树，并对结果简单分析

```python
from sklearn.tree import export_graphviz
import pydotplus

#export_graphviz(grid_search.best_estimator_, out_file='makemoons.dot')
label = x.columns
data_class = ['0','1']
dot_data = export_graphviz(grid_search.best_estimator_, out_file=None, feature_names=label, class_names=data_class,
                               filled=True, rounded=True, special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data)
graph.write_pdf('titanic.pdf')
f = open('DecisionTree.png', 'wb')
f.write(graph.create_png())
f.close()
```
